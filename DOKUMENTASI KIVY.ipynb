__pengertian kivy__

Kivy adalah GUI/NUI framework yang menggunakan bahasa pemrograman Python, dapat dijalankan pada Windows, Linux, OS X, Android dan iOS dengan menggunakan code yang sama dan menggunakan MIT license sehingga 100% free to use dan boleh dikomersialisasikan. Kivy didesain untuk membantu pembuatan dan pengembangan aplikasi cross-platform secara cepat.

__CARA INSTALASI KIVY__

*pastikan terhubung ke internet

1. Pada tampilan cmd silahkan anda ketik perintah dibawah ini lalu tekan enter

python -m pip install --upgrade pip wheel setuptools

2. Jika langkah pertama sudah selesai ketik perintah berikut pada cmd pula, tunggu hingga proses selesai

4. Langkah terakhir
python -m pip install kivy

# __Penjelasan Class__

__1. Class CalculatorWidget__



Pada Class CalculatorWidget terdapat 5 method yaitu :

1. calc_error
2. calculate
3. delete
4. clear
5. switch

Pada method __calc_error__ berguna untuk menampilkan pesan error jika ada..
dan pada method calc_error juga berguna untuk membuka popup / membuka aplikasi calculatornya 

Pada method __calculate__ berguna untuk menampung input yang mau di operasikan

pada method __delete__ berguna untuk menghapus segala bentuk input yang ada di calculator

pada method __clear__ berguna untuk menghilangkan semua input yang ada di layar calculator.. Bedanya dengan method __delete__ adalah jika __delete__ hanya menghapus 1 persatu tetapi jika method __clear__ langsung menghapus segala maccam input yang ada 

pada method __switch__ berguna untuk menghitung operasi yang sudah tertampung pada method __calculate__


__2. Class CalculatorApp__



Pada class CalculatorAPP terdapat 1 method saja yaitu :

1. build

Pada method __build__ berguna untuk mensetting aplikasi calculatornya dan disitu terdapat return class CalculatorWidget dan pada bagian bawah sendiri ada method untuk memanggil class calculatorApp
